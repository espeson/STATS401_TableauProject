```python
import numpy as np
import pandas as pd
import json
import jieba
from tqdm.notebook import tqdm

# Custom format with ANSI escape codes for dark green color
dark_green = "\033[1;32;40m"
bar_format = f"{dark_green}{{l_bar}}{{bar:50}} [{{elapsed}}]{{r_bar}}"
```

```python
records = []
with open("tweet_spider_by_keyword_20230905090130.jsonl", "r") as f:
    while line := f.readline():
        records.append(json.loads(line))

with open("tweet_spider_by_keyword_20230905085828.jsonl", "r") as f:
    while line := f.readline():
        records.append(json.loads(line))

with open("tweet_spider_by_keyword_20230905204720.jsonl", "r") as f:
    while line := f.readline():
        records.append(json.loads(line))
```

```python
user_table = {}
user_table_columns = list(records[0]["user"].keys())
for i in user_table_columns:
    user_table[i] = []

for i in records:
    for j in user_table_columns:
        try:
            user_table[j].append(i["user"][j])
        except:
            user_table[j].append(np.nan)

user_table = pd.DataFrame(columns=user_table_columns, data=user_table)
user_table = user_table.rename(columns={"_id": "user_id"})
user_table = user_table.set_index("user_id")
display(user_table)
```

```python
cn_prov = []
with open("Chinese_provience.csv", "r", encoding="utf-8") as f:
    prov_name = ""
    while prov_name := f.readline():
        cn_prov.append(prov_name[:-1])
display(cn_prov)
```

```python
post_table = {}
post_table_columns = list(records[0].keys())
post_table_columns.append("country")
for i in post_table_columns:
    post_table[i] = []
#print(post_table_columns)

bar = tqdm(total=len(records), bar_format=bar_format)
for i in records:
    for j in post_table_columns:
        if j == "country":
            continue
        try:
            if j == "user":
                post_table[j].append(i["user"]["_id"])
            elif j == "ip_location":
                if i[j] is None:
                    post_table["country"].append(np.nan)
                    post_table["ip_location"].append(np.nan)
                    continue                   
                location = i[j][4:]
                if location in cn_prov:
                    post_table["country"].append("China")
                    post_table["ip_location"].append(location)
                elif location == "中国台湾":
                    post_table["country"].append("Taiwan")
                    post_table["ip_location"].append(np.nan)
                elif location == "中国香港":
                    post_table["country"].append("China")
                    post_table["ip_location"].append(location[2:])
                elif location == "中国澳门":
                    post_table["country"].append(location[2:])
                    post_table["ip_location"].append(np.nan)
                else:
                    post_table["country"].append(location)
                    post_table["ip_location"].append(np.nan)
            else:
                post_table[j].append(i[j])
        except FileExistsError:
            post_table[j].append(np.nan)
    bar.update()
bar.close()

post_table = pd.DataFrame(columns=post_table_columns, data=post_table)
post_table = post_table.rename(columns={"_id": "post_id", "user": "user_id"})
post_table = post_table.set_index("post_id")
display(post_table)
```

```python
token_table = {"post_id": [], "token": []}
stop_words = []
with open("stop_words.csv", "r") as f:
    while line := f.readline():
        stop_words.append(line[:-1])
stop_words.extend(["\n", " ", "\""])

bar = tqdm(total=len(post_table.index), bar_format=bar_format)
for row_index, post in post_table.iterrows():
    token_list = jieba.lcut_for_search(post[8])
    for token in token_list:
        if token not in stop_words:
            token_table["post_id"].append(row_index)
            token_table["token"].append(token)
    bar.update()
bar.close()

token_table = pd.DataFrame(data=token_table)
display(token_table)
```

```python
with pd.ExcelWriter(
    "table.xlsx",
    mode="w",
    engine="openpyxl",
    #if_sheet_exists="replace",
) as writer:
    post_table.to_excel(writer, sheet_name="posts")
    user_table.to_excel(writer, sheet_name="users")
```

```python
from argostranslate import translate
from argostranslate import package

package.install_from_path('translate-zh_en-1_7.argosmodel')

def get_argos_model(source, target):
    lang = f'{source} -> {target}'
    source_lang = [model for model in translate.get_installed_languages() if lang in map(repr, model.translations_from)]
    print(source_lang)
    target_lang = [model for model in translate.get_installed_languages() if lang in map(repr, model.translations_to)]
    print(target_lang)
    
    return source_lang[0].get_translation(target_lang[0])

argos_zh_en = get_argos_model('Chinese', 'English')
```

```python
# Processed Token Table
processed_token_table = {}
for i in token_table["token"]:
    try:
        processed_token_table[i] += 1
    except:
        processed_token_table[i] = 1

_processed_token_table = {"Token": [], "Count": [], "Token_en": []}
bar = tqdm(total=len(processed_token_table))
for ind, (key, value) in enumerate(processed_token_table.items()):
    if ind >= 1000:
        break
    if value <= 150:
        bar.update()
        continue
    _processed_token_table["Token"].append(key)
    _processed_token_table["Token_en"].append(argos_zh_en.translate(key))
    _processed_token_table["Count"].append(value)
    bar.update()
bar.close()

display(len(_processed_token_table["Token"]))

_processed_token_table_df = pd.DataFrame(data = _processed_token_table)
_processed_token_table_df.to_csv("processed_token_table.csv")
with open("processed_token_table.json", "w", encoding="utf-8") as f:
    json.dump(_processed_token_table, f, indent=2)
```
